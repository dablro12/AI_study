### 리스트로 배열 만들기


image.png

```
#리스트로 배열만들기 np.array(리스트)
a1 = np.array(list(range(5)))
print(a1)
print(type(a1)) #-> type(var) : ndarray : n차원의 array임을 알 수 있다.
print(a1.shape) #-> var.shape : array에 대한 모양을 볼 수 있다.
print(a1.ndim) #-> var.ndim : array에 대한 차원을 알 수 있다.
print(a1[0],a1[1],a1[2],a1[3],a1[4])

#array 안의 요소를 바꿀 수 있다.
a1[0] = 5
a1[1] = 4
a1[2] = 3
a1[3] = 2
a1[4] = 1
print(a1)
```

```
#2차원을 알아보자
a2 = np.array([list(range(1,4)),list(range(4,7)),list(range(7,10))])
print(type(a2)) #-> type(var) : ndarray : n차원의 array임을 알 수 있다.
print(a2.shape) #-> var.shape : array에 대한 모양을 볼 수 있다.
print(a2.ndim) #-> var.ndim : array에 대한 차원을 알 수 있다. 
print(a2)

print(a2[0,0], a2[1,1], a2[2,2])
```

```
a3 = np.array([
    [list(range(1,4)),list(range(4,7)),list(range(7,10))],
     [list(range(1,4)),list(range(4,7)),list(range(7,10))],
     [list(range(1,4)),list(range(4,7)),list(range(7,10))] 
     ])

print(type(a3)) #-> type(var) : ndarray : n차원의 array임을 알 수 있다.
print(a3.shape) #-> var.shape : array에 대한 모양을 볼 수 있다.
print(a3.ndim) #-> var.ndim : array에 대한 차원을 알 수 있다. 
print(a3)
```

### 배열 생성 및 초기화

* `zeros()`: 모든 요소를 0으로 초기화

```
a = np.zeros((3,3))
a
```

* `ones()`: 모든 요소를 1로 초기화

```
a = np.ones((3,3,3))
a
```

* `full()`: 모든 요소를 지정한 값으로 초기화

```
a = np.full((3,3), 1.23)
a
```

* `eye()`: 단위행렬(identity matrix) 생성
  + 주대각선의 원소가 모두 1이고 나머지 원소는 모두 0인 정사각 행렬

```
#단위행렬은 주대각선 원소가 모두 1이고 나머지 원소는 모두 0인 rectangular matrix
a = np.eye(3)
a
```

* `tri()`: 삼각행렬 생성

```
a = np.tri(3)
a
```

* `empty()`: 초기화되지 않은 배열 생성
  + 초기화가 없어서 배열 생성비용 저렴하고 빠름
  + 초기화되지 않아서 기존 메모리 위치에 존재하는 값이 있음

```
b = np.empty((3,3))
b
#empty는 초기화가 없어 배열을 만드는게 쉽지만 기존 메모리에 위치하는 값(b 대신 a)를 넣으면 기존의 값이 나온다.
```

* `_like()`: 지정된 배열과 shape가 같은 행렬 생성
  + `np.zeros_like()`
  + `np.ones_like()`
  + `np.full_like()`
  + `np.empty_like()`

```
#np.zeros_like(var)
print(a1)
print(np.zeros_like(a1)) #a1과 같은 shape의 0으로 이루어진 배열을 만들어줌
```

```
print(a2)
print(np.ones_like(a2)) #a2와 같은 shape으로 1로 이루어진 배열을 만들어줌
```

```
print(a3)
#np.full_like(var, 특정값)
print(np.full_like(a3, 10)) 
```

### 생성한 값으로 배열 생성

* `arange()`: 정수 범위로 배열 생성

```
#np.arange()리스트에서 자주사용하는 range()와 같다.
np.arange(0,4,2)
```

* `linspace()`: 범위 내에서 균등 간격의 배열 생성

```
#np.linspace()
np.linspace(0,1,5) #0~1사이를 5개로 나눠서 배열로 만듬
```

* `logspace()`: 범위 내에서 균등간격으로 로그 스케일로 배열 생성

```
#np.logspace() log스케일로 배열을 생성
np.linspace(0.01, 1, 4)
```

### 랜덤값으로 배열 생성


* `random.random()`: 랜덤한 수의 배열 생성

```
#np.random.random(shape) 실수값을 0~1사이에서 반환
np.random.random((3,3))
```

* `random.randint()`: 일정 구간의 랜덤 정수의 배열 생성

```
np.random.randint(0,10, (3,3)) #0~10사이의 값을 정수형으로 반환
#꼭 range를 정해주어야함
```

* `random.normal()`: 정규분포(normal distribution)를 고려한 랜덤한 수의 배열 생성
* 평균=0, 표준편차=1, 3 x 3 배열

```
#np.random.normal(range, shape)
np.random.normal(0,10, (3,3))
```

* `random.rand()`: 균등분포(uniform distribution)를 고려한 랜덤한 수의 배열 생성

```
#np.random.rand(shape)
np.random.rand(3,3)
```

* `random.randn()`: 표준 정규 분포(standard normal distribution)를 고려한 랜덤한 수의 배열 생성

```
#np.radndom.randn(3,3) #표준정규 분포를 고려한 수라는 것을 알아야함
#rand vs randn
np.random.randn(3,3)
```
