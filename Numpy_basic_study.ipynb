### 배열 생성 및 초기화

* `zeros()`: 모든 요소를 0으로 초기화

```
a = np.zeros((3,3))
a
```

* `ones()`: 모든 요소를 1로 초기화

```
a = np.ones((3,3,3))
a
```

* `full()`: 모든 요소를 지정한 값으로 초기화

```
a = np.full((3,3), 1.23)
a
```

* `eye()`: 단위행렬(identity matrix) 생성
  + 주대각선의 원소가 모두 1이고 나머지 원소는 모두 0인 정사각 행렬

```
#단위행렬은 주대각선 원소가 모두 1이고 나머지 원소는 모두 0인 rectangular matrix
a = np.eye(3)
a
```

* `tri()`: 삼각행렬 생성

```
a = np.tri(3)
a
```

* `empty()`: 초기화되지 않은 배열 생성
  + 초기화가 없어서 배열 생성비용 저렴하고 빠름
  + 초기화되지 않아서 기존 메모리 위치에 존재하는 값이 있음

```
b = np.empty((3,3))
b
#empty는 초기화가 없어 배열을 만드는게 쉽지만 기존 메모리에 위치하는 값(b 대신 a)를 넣으면 기존의 값이 나온다.
```

* `_like()`: 지정된 배열과 shape가 같은 행렬 생성
  + `np.zeros_like()`
  + `np.ones_like()`
  + `np.full_like()`
  + `np.empty_like()`

```
#np.zeros_like(var)
print(a1)
print(np.zeros_like(a1)) #a1과 같은 shape의 0으로 이루어진 배열을 만들어줌
```

```
print(a2)
print(np.ones_like(a2)) #a2와 같은 shape으로 1로 이루어진 배열을 만들어줌
```

```
print(a3)
#np.full_like(var, 특정값)
print(np.full_like(a3, 10)) 
```

### 슬라이싱(Slicing)

* 슬라이싱 구문: `a[start:stop:step]`
* 기본값: start=0, stop=ndim, step=1

```
a1 = np.array([1,2,3,4,5])
print(a1[0:2])
print(a1[0:])
print(a1[:1])
print(a1[::2]) #index : 0, 2, 4인 값인 1,3,5 나옴가 나옴
print(a1[::-1]) #뒤에서부터 출력
```

```
a2 = np.random.rand(3,3)
print(a2)
print()
print(a2[1])
print(a2[1, :])
print(a2[:2, :2])
print()
print(a2[::-1, ::-1 ]) #뒤집어서 출력
```

### 불리언 인덱싱(Boolean Indexing)

```
print(a1)
b1 = [False, True, False, True, False]
print(a1)
print(a1[b1]) #index값 대신에 b1을 넣어주면 True위치에 있는 값들만 조회할 수 있음

b1 = [True, True, False, False, False]
print(a1)
print(a1[b1])
```

```
print(a2)
b1 = np.random.randint(0, 2, (3,3), dtype=bool)
print(b1)
print(a2[b1]) #True로 indexing된 부분만 가지고 온다. 
```

## 배열 값 삽입/수정/삭제/복사 46:00~

### 배열 값 삽입

* `insert()`: 배열의 특정 위치에 값 삽입
* axis를 지정하지 않으면 1차원 배열로 변환
* 추가할 방향을 axis로 지정
* 원본 배열 변경없이 새로운 배열 반환

```
print(a1)
#b1을 하나만든다.
#np.insert(var, index, object)
b1 = np.insert(a1, 0, 10)
print(b1)
print(a1) #원본이 변경되지 않는다.

c1 = np.insert(a1, 2, 10)
print(c1) #-> 3,4,5는 기존 idx + 1로 밀려난다.
```

```
print(a2)
b2 = np.insert(a2, 1, 10, axis = 1) #2차원이므로 축을 지정할 수 있다. axis = 차원 (0~ndim)
print(b2)
c2 = np.insert(a2, 2, 10, axis = 0)
print(c2)
```

### 배열 값 수정

* 배열의 인덱싱으로 접근하여 값 수정

```
print(a1)
a1[0] = 5
a1[1] = 4
print(a1)
print()
a1[:2] = 9 #slicing을 이용해 idx 0, 1만 9로 바꿈
print(a1)

i = np.array([1,3,4]) #fancy slicing 할 수 있음
a1[i] = 0 
print(a1)
a1[i] += 4 #idx = 1,3,4에 9를  더함
print(a1)
```

```
print(a2)
a2[0,0] = 1
a2[1,1] = 2
a2[2,2] = 3
a2[0] = 4 #하나만 쓰면 해당 row 전체 요소가 4가 됨
print(a2)
a2[1:, 2] = 9
print(a2)
row = np.array([0,1])
col = np.array([1,2])
a2[row,col] = 10
print(a2)
```

### 배열 값 삭제

* `delete()`: 배열의 특정 위치에 값 삭제
* axis를 지정하지 않으면 1차원 배열로 변환
* 삭제할 방향을 axis로 지정
* 원본 배열 변경없이 새로운 배열 반환

```
#delete 삭제된 새로운 배열이 return 기존의 값은 변하지 않음
print(a1)
b1 = np.delete(a1, 1) #idx = 1번쨰 값을 삭제
print(b1)
print(a1) #a1의 값은 변경되지 않는다.
```

```
print(a2)
b2 = np.delete(a2, 1, axis = 0) #0차원의 array(row) 의 idx = 1번쨰 값들을 제거 
print(b2)
b2 = np.delete(a2, 1)
print(b2)
c2 = np.delete(a2, 0, axis = 1) #1차원 = col의 idx =0번쨰 값들을 제거
print(c2)
```

### 배열 복사

* 리스트 자료형과 달리 배열의 슬라이스는 복사본이 아님

```
print(a2)
print(a2[:2, :2])
a2_sub = a2[:2, :2]
print(a2_sub)
a2_sub[:, 1] = 3
print(a2_sub)
print(a2) #-> 원본 배열도 바뀜! -> 받아온 어떠한 배열도 슬라이싱이 됨
```


* `copy()`: 배열이나 하위 배열 내의 값을 명시적으로 복사

```
#기존의 값들을 벼녕하지 않고 진행하고 싶다면 var.copy() 명령어를 붙여햐 한다.
print(a2)
print(a2[:2, :2])
a2_sub = a2[:2, :2].copy()
print(a2_sub)
a2_sub[:, 1] = 3
print(a2_sub)
print(a2) #-> 원본 배열 a2가 바뀌지않음
```

### 배열 분할

## 배열 연산

* NumPy의 배열 연산은 벡터화(vectorized) 연산을 사용
* 일반적으로 NumPy의 범용 함수(universal functions)를 통해 구현

### 브로드캐스팅(Broadcasting)

image.png

image.png

image.png

```
#numpy 배열은 벡터화한 변수 연산을 사용하므로 반복적인 계산을 효율적으로 수행할 수 있다.
#범용함수(내적 함수, invertion 함수 등)를 쓸 수 있어서 좋음

#브로드캐스팅이란 :  1,2,3 이라는 array에서 +5를 하면 같은 shape의 5라는 요소를 가진 1*3의 array를 더할 수 있다.
#즉 부족한 부분을 채워서 연산할 수 있게 해줌
```

```
a1 = np.array([1,2,3]) #첫번쨰 그림을 보고 똑같이 만듬
print(a1)
print(a1+5)
print()

#2번쨰그림
a2 = np.arange(1,10).reshape(3,3) 
print(a2)
print(a2+a2)

#3번쨰그림
b2 = np.array([1,2,3]).reshape(3,1)
print(b2)
print(a1+b2)

```

### 산술 연산(Arithmetic Operators)

image.png

```
a1 = np.arange(1,10) 
print(a1)

#범용함수를 써보자
#np.add(var, 더할 숫자)
print(a1 + 1) #각 요소에 1을 더한 값이 나옴 
print(np.add(a1, 10))

print()
#np.subtract(var, 뺼 숫자)
print(a1 -2)
print(np.subtract(a1,10))

print()
#np.negative(var) : -를 붙힘
print(-a1)
print(np.negative(a1))

print()
#np.multiply(var, 곱할 숫자)
print(a1*2) 
print(np.multiply(a1, 2)) 

print()
#np.divide(var, 나눌 숫자)
print(a1 / 2)
print(np.divide(a1, 2))

print()
#np.floor_divide(var, 숫자)나눗셈 내림  // 역할
print(a1 // 2)
print(np.floor_divide(a1, 2))

print()
# np.power(var, 숫자) : 제곱
print(a1 ** 2)
print(np.power(a1, 2))

print()
#np.mod(var, 숫자) : 나머지값 구하기
print(a1 % 2)
print(np.mod(a1, 2))
```

```
a1 = np.arange(1,10)
print(a1)
b1 = np.random.randint(1,10, size = 9)
print(b1)
print(a1 + b1)
print(a1 - b1)
print(a1 * b1)
print(a1 / b1)
print(a1 // b1)
print(a1 ** b1)
print(a1 % b1)
```

```
#2차원 array도 해보자
a2 = np.arange(1,10).reshape(3,3)
print(a2)
b2 = np.random.randint(1,10, size= (3,3))
print(b2)

print(a2 + b2)
print(a2 - b2)
print(a2 * b2)
print(a2 / b2)
print(a2 // b2)
print(a2 ** b2)
print(a2 % b2)
```

### 집계 함수(Aggregate Functions)

image.png

#### sum(): 합 계산

```
#np.sum(var) 또는 var.sum(), optin : axis 축을 지정해줄 수 있음

a2 = np.random.randint(1,10, size=(3,3))
print(a2)

print(a2.sum(), np.sum(a2, axis = 0))
print(a2.sum(axis = 1), np.sum(a2, axis = 0)) 
```

#### cumsum(): 누적합 계산

```
print(a2)
print(np.cumsum(a2, axis = 0))
print(np.cumsum(a2, axis = 1)) 
```

#### diff(): 차분 계산

```
#미분
print(a2)
print(np.diff(a2)) #2에서 3이 되기 위해 1이 필요하므로 차분해준다.
print(np.diff(a2, axis = 0))
print(np.diff(a2, axis = 1))
```

#### prod(): 곱 계산

```
#곱계산
print(a2)
print(np.prod(a2))
print(np.prod(a2, axis = 0))
print(np.prod(a2, axis = 1))
```

#### cumprod(): 누적곱 계산

```
#누적곱계산
print(a2)
print(np.cumprod(a2))
print(np.cumprod(a2, axis = 0))
print(np.cumprod(a2, axis = 1))
```

#### dot()/matmul(): 점곱/행렬곱 계산

```
#행렬곱 ! 중요! np.matmul(var)
b2 = np.ones_like(a2)

print(a2)
print(b2)
print()
print(np.dot(a2,b2))
print(np.matmul(a2,b2))
```

#### tensordot(): 텐서곱 계산

```
print(a2)
print(b2)
print()
print(np.tensordot(a2,b2))
print()
print(np.tensordot(a2,b2, axes = 0))
print()
print(np.tensordot(a2,b2, axes = 1))
```

#### cross(): 벡터곱

### 비교 연산(Comparison Operators)


image.png

```
a1 = np.arange(1,10).reshape(3,3)
print(a1)
print(a1 == 5) 
print(a1 != 5) 
print(a1 < 5) 
print(a1 > 5) 
print(a1 <= 5) 
print(a1 >= 5) 
```

```
a2 = np.arange(1,10).reshape(3,3)
print(a2)
print(np.sum(a2))
print(np.count_nonzero(a2 > 5))
print(np.sum(a2>5))
print(np.sum(a2>5, axis= 0))
print(np.sum(a2>5, axis= 1))
```

image.png

```
a1 = np.array([1,2,3,4,5])
b1 = np.array([1,2,3,4,3])

print(a1)
print(b1)
print(np.isclose(a1,b1))
```

```
a1 = np.array([np.nan, 2, np.inf, 4, np.NINF])
print(a1)
print(np.isnan(a1))
print(np.isfinite(a1))
print(np.isinf(a1))
```

#### 불리언 연산자(Boolean Operators)


image.png

```
a2 = np.arange(1,10).reshape(3,3)
print(a2)

print((a2>5) & (a2 < 8))
print(a2[(a2>5) & (a2 < 8)])


print((a2>5) | (a2 < 8))
print(a2[(a2>5) | (a2 < 8)])


print((a2>5) ^ (a2 < 8))
print(a2[(a2>5) ^ (a2 < 8)])


print(~ (a2 < 8))
print(a2[~ (a2 < 8)])

```

### 배열 정렬

```
a1 = np.random.randint(1,10, size =10)
print(a1)
print(np.sort(a1))
print(a1)
print(np.argsort(a1)) #정렬되는 요소 index값만 반환
print(a1) #-> 바뀌지 않는다.
print(a1.sort()) #-> 이건 바뀐다. None을 반환하고 a1을 바꿈
print(a1)
```

```
a2 = np.random.randint(1,10, size= (3,3))
print(a2)
print(np.sort(a2, axis= 0))
print(np.sort(a2, axis= 1))
```

#### 부분 정렬

* `partition()`: 배열에서 k개의 작은 값을 반환

```
a1 = np.random.randint(1,10, size = 10)
print(a1)
print(np.partition(a1, 3)) #작은 값 3개를 반환 하고 나머지를 뒤로 보냄
```

```
#N차원도 가능
a2 = np.random.randint(1,10, size=(5,5))
print(a2) 
print(np.partition(a2, 3))
print(np.partition(a2, 3,axis = 0))
print(np.partition(a2, 3,axis = 1))
```
