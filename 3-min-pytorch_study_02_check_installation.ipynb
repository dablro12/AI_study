```python
from importlib import import_module 

required_package = [ 
    ('torch', '파이토치'),
    ('torchvision', '토치비젼'),
    ('torchtext', '토치텍스트'),
    ('numpy', '넘파이'),
    ('matplotlib', '맷플럽랍'),
    ('sklearn', '사이킷런')
]

installed_packages = []
uninstalled_packages = []

def run():
  is_ready = True
  for package_name, korean_name in required_package:
    printed_name = "%s(%s)" %(korean_name, package_name)
    try:
      imported_package = import_module(package_name)
      installed_packages.append("%s 버전 : %s" % (printed_name, imported_package.__version__))
    except:
      is_ready = False
      uninstalled_packages.append(printed_name)

    if is_ready == True:
      print("pytorch env all installed")
      print("installed library list")
      for pkg in installed_packages:
        print(" * " + pkg)
    else:
      print("uninstalled library is exist")
      print("installed library list")
      for pkg in installed_packages:
        print(" * " + pkg)
      print("uninstalled library info")
      for pkg in uninstalled_packages:
        print(" * " + pkg)
    return is_ready 


if __name__ == "__main__":
  run()
      
      
      
```
